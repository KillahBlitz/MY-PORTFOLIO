```python
import requests
import pandas as pd
```

```python
#obtener y conectar con los datos del API
url = "https://pokeapi.co/api/v2/pokemon/pikachu"
response = requests.get(url)
data = response.json()
#acceder al nombre
nombre = data['name']
print(f"Nombre: {nombre}")
#acceder a las estadisticas
stats = data['stats']
for stat in stats:
    stat_name = stat['stat']['name']
    base_stat = stat['base_stat']
    print(f"{stat_name.capitalize()}: {base_stat}")
#acceder a los tipos
types = data['types']
for tipo in types:
    type_name = tipo['type']['name']
    print(f"Tipo: {type_name}")
```

```python
#convertir estadisticas en data frame
stats_df = pd.DataFrame([{
    'stat': stat['stat']['name'],
    'value': stat['base_stat']
} for stat in stats])

print("Estadisticas de Pikachu:")
display(stats_df)
```

```python
##procesamientos complejos
##traer datos de varios pokemones
pokemons = ['pikachu','bulbasaur','charmander']
all_stats = []
for pokemon in pokemons:
    url = f"https://pokeapi.co/api/v2/pokemon/{pokemon}"
    response = requests.get(url)
    data = response.json()

    #acceder a las estadisticas
    for stat in data['stats']:
        all_stats.append({
            'pokemon': data['name'],
            'stat': stat['stat']['name'],
            'value': stat['base_stat']
        })

print(all_stats)
```

```python
#conversion a data frame
all_stats_df = pd.DataFrame(all_stats)
display(all_stats_df)
```

```python
#mejorar el data frame con pivotes
pivot_df = all_stats_df.pivot_table(index='pokemon', columns='stat', values='value', aggfunc='mean')
display(pivot_df)
```
