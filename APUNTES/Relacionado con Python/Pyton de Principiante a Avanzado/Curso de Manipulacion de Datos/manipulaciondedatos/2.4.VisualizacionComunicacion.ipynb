```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno
```

```python
df = pd.read_csv('datos_con_faltantes.csv')
display(df)
```

```python
#mapa de calor de datos faltantes
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Mapa de calor de datos faltantes')
plt.show()
```

```python
#generar matriz de datos faltantes
msno.matrix(df)
plt.title('Matriz de datos faltantes')
plt.show()
```

```python
#Grafico de barras
df.isnull().sum().plot(kind='bar')
plt.title('Cantidad de datos faltantes por columna')
plt.xlabel('Columnas')
plt.ylabel('Numero de datos faltantes')
plt.show()
```

```python
#mapa de coorrelacion de datos faltantes
msno.heatmap(df)
plt.title('Mapa de correlacion de datos faltantes')
plt.show()
```
