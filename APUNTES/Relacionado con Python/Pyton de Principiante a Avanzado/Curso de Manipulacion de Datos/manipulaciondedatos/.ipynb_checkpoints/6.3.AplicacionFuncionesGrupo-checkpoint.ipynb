```python
import pandas as pd
```

```python
#data frame de ejemplo
data = {
    'producto': ['Manzana', 'Manzana', 'Banana', 'Banana', 'Cereza', 'Cereza',  'Manzana', 'Banana', 'Cereza'],
    'categoria': ['Fruta', 'Fruta', 'Fruta', 'Fruta', 'Fruta', 'Fruta', 'Fruta', 'Fruta', 'Fruta'],
    'ventas': [50, 30, 20, 15, 10, 25, 45, 55, 40],
    'precio': [0.5, 0.5, 0.3, 0.3, 0.8, 0.8, 0.5, 0.3, 0.8],
    'fecha': pd.date_range(start='2024-01-01', periods=9, freq='D')
}

df = pd.DataFrame(data)
display(df)
```

```python
resumen_de_ventas_por_producto = df.groupby('producto')['ventas'].agg(['sum','mean']).reset_index()
display(resumen_de_ventas_por_producto)
```

```python
def rango_ventas(grupo):
    return grupo.max() - grupo.min()

rango_por_producto = df.groupby('producto')['ventas'].agg(rango_ventas).reset_index()
display(rango_por_producto)
```

```python
df['ventasnormalizadas'] = df.groupby('producto')['ventas'].transform(lambda x: (x - x.mean()) / x.std())
display(df)
```

```python
df['ingresototal'] = df['ventas'] * df['precio']
df_con_ingreso_total = df.groupby(['producto', 'categoria']).agg({
    'ventas': 'sum',
    'ingresototal': 'sum'
}).reset_index()
display(df)
display(df_con_ingreso_total)
```
