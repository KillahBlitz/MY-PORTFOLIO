```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'producto': ['Manzana', 'Banana', 'Cereza', 'Durazno', 'Pera'],
    'categoria': ['Fruta', 'Fruta', 'Fruta', 'Fruta', 'Fruta'],
    'calidad': ['Alta', 'Media', 'Baja', 'Alta', 'Media'],
    'ventas': [50,30,70,85,40]
}

df = pd.DataFrame(data)
display(df)
```

```python
#codificacion ordinal
mapacalidad = {'Baja': 1, 'Media': 2, 'Alta': 3} #mapa numerico de la categorizacion

#mapeo de codificacion ordinal
df['calidadordinal'] = df['calidad'].map(mapacalidad)
display(df)
```

```python
#codificacion one hot
df_one_hot = pd.get_dummies(df, columns=['calidad'])
display(df_one_hot)
```

```python
#visualizar transformaciones
plt.bar(df['producto'], df['calidadordinal'])
plt.title("Codificacion Ordinal de calidad")
plt.show()
```
