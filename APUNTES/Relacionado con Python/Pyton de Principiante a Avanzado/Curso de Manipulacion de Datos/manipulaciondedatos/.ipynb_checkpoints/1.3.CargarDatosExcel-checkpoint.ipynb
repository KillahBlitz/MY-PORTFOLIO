```python
import pandas as pd
```

```python
#ocupamos openpyxl para leer archivos excel
#de esta forma podemos ver por default la hoja1, pero con sheet_name podemos seleccionar la hoja
df_excel = pd.read_excel('datos_ejemplo.xlsx', sheet_name='Hoja1')
display(df_excel)
```

```python
#hay que cersiorarnos de las hojas que tenemos y sus nombres, podemos consultarlos con:
xls = pd.ExcelFile('datos_ejemplo.xlsx') #aqui inspeccionamos el archivo, no el contenido
print(xls.sheet_names)
```

```python
#para ver el contenido de todas las hojas
#bucle for:
for sheet_name in xls.sheet_names:
    df = pd.read_excel(xls, sheet_name=sheet_name)
    print(f"Contenido de la hoja: {sheet_name}")
    display(df.head())
```
