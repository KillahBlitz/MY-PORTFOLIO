```python
import pandas as pd
```

```python
df_duplicated = pd.read_csv('DatosDuplicados.csv')
display(df_duplicated)
```

```python
#eliminar datos duplicados
df_no_duplicated = df_duplicated.drop_duplicates()
display(df_no_duplicated)
```

```python
#eliminacion selectiva por columnas especificas
df_name = df_duplicated.drop_duplicates(subset=['Nombre'], keep='last')
display(df_name)
```

```python
#sumarizacion de duplicados
df_summarized = df_duplicated.groupby('Nombre').agg({
    'Edad': 'first',
    'Salario': 'mean',
    'Fecha_Ingreso': 'first'
}).reset_index()

display(df_summarized)
```

```python
#identificar los duplicados
df_duplicated['Es_Duplicado'] = df_duplicated.duplicated(subset=['Nombre'])
display(df_duplicated)
```

```python
#fusionar duplicados con nuevos datos
df_nuevo = pd.DataFrame({
    'ID': [1,2,11],
    'Nombre': ['Ana','Luis','Gabriela'],
    'Edad': [24,36,27],
    'Salario': [51000,61000,68000],
    'Fecha_Ingreso': ['2021-01-15','2021-06-15','2021-08-01']
})
display(df_nuevo)
```

```python
#concatenar dos dataframes
df_merged = pd.concat([df_duplicated, df_nuevo]).drop_duplicates(subset=['ID'], keep='last')
display(df_merged)
```
