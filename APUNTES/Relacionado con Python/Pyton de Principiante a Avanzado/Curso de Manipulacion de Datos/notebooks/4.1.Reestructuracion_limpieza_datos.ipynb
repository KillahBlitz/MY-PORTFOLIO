```python
import pandas as pd
import numpy as np
```

```python
data = {
    'ID': range(1, 21),
    'Nombre': ['Ana', 'Luis', 'Carlos', 'María', 'Elena', 'Jorge', 'Sofía', 'Andrés', 'Laura', 'Pedro',
               'Paula', 'Diego', 'Carmen', 'Marta', None, 'Pablo', 'Lucía', 'Manuel', 'Carla', 'David'],
    'Edad': [23, 35, 30, 22, 28, np.nan, 34, 45, 29, 41, 38, 25, 33, 27, 32, 36, np.nan, 39, 31, 40],
    'Salario': [50000, 60000, np.nan, 58000, 49000, 62000, 72000, 61000, 52000, 67000,
                55000, 53000, np.nan, 69000, 64000, 71000, 68000, 73000, np.nan, 66000],
    'Fecha_Ingreso': ['2020-01-10', '2019-06-15', '2020-03-25', None, '2018-07-30', '2017-09-14',
                      '2016-11-22', '2021-04-01', '2015-05-19', '2014-08-25', '2013-12-30', '2012-02-11',
                      '2011-10-17', None, '2020-09-09', '2018-04-05', '2019-11-22', '2010-07-27', '2020-05-13', '2017-01-29']
}
df = pd.DataFrame(data)
display(df)
```

```python
df.isnull().sum()
```

```python
df.fillna({'Salario': df['Salario'].mean()}, inplace=True)

display(df)
```

```python
df.dropna(subset=['Nombre', 'Fecha_Ingreso'], inplace=True)
display(df)
```

```python
df['Rango_Edad'] = pd.cut(df['Edad'], bins=[20, 30, 40, 50], labels=['20-30', '30-40', '40-50'])
df_grouped = df.groupby('Rango_Edad')['Salario'].mean()
display(df_grouped)
```

```python
df_grouped = df.groupby('Rango_Edad', observed=False)['Salario'].mean()
display(df_grouped)
```

```python
df_grouped = df.groupby('Rango_Edad', observed=True)['Salario'].mean()
display(df_grouped)
```
