```python
import pandas as pd
```

```python
data = {
    'producto': ['Manzana', 'Manzana', 'Banana', 'Banana', 'Cereza', 'Cereza', 'Manzana', 'Banana', 'Cereza'],
    'categoría': ['Fruta', 'Fruta', 'Fruta', 'Fruta', 'Fruta', 'Fruta', 'Fruta', 'Fruta', 'Fruta'],
    'ventas': [50, 30, 20, 15, 10, 25, 45, 55, 40],
    'precio_unitario': [0.5, 0.5, 0.3, 0.3, 0.8, 0.8, 0.5, 0.3, 0.8],
    'fecha': pd.date_range(start='2024-01-01', periods=9, freq='D')
}

df = pd.DataFrame(data)
display(df)
```

```python
def calcular_ingreso_total(row):
    return row['ventas'] * row['precio_unitario']

df['ingreso_total'] = df.apply(calcular_ingreso_total, axis=1)
display(df[['producto', 'ventas', 'precio_unitario', 'ingreso_total']])
```

```python
def clasificar_producto(ventas):
    if ventas > 40:
        return 'Alto'
    elif 20 <= ventas <= 40:
        return 'Medio'
    else:
        return 'Bajo'

df['clasificación'] = df['ventas'].apply(clasificar_producto)
display(df[['producto', 'ventas', 'clasificación']])
```

```python
df['producto_upper'] = df['producto'].apply(lambda x: x.upper())
display(df)
```

```python
def calcular_porcentaje(ventas):
    total = ventas.sum()
    return ventas / total * 100

df['porcentaje_ventas'] = df.groupby('producto')['ventas'].transform(calcular_porcentaje)
display(df[['producto', 'ventas', 'porcentaje_ventas']])
```
