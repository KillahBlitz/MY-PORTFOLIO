```python
import pandas as pd
```

```python
df =  pd.read_csv("datos_con_faltantes.csv")
display(df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Mapa de calor de datos faltantes')
plt.show()
```

```python
import missingno as msno
```

```python
msno.matrix(df)
plt.title('Matriz de datos faltantes')
plt.show()
```

```python
df.isnull().sum().plot(kind='bar')
plt.title('Cantidad de datos faltantes por columna')
plt.xlabel('Columnas')
plt.ylabel('Número de datos faltantes')
plt.show()
```

```python
msno.heatmap(df)
plt.title('Mapa de correlación de datos faltantes')
plt.show()
```
