```python
import pandas as pd
```

```python
data_largo = {
    'Producto': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Mes': ['Enero', 'Febrero', 'Enero', 'Febrero', 'Enero', 'Febrero'],
    'Ventas': [100, 150, 200, 250, 300, 350]
}

df_largo = pd.DataFrame(data_largo)
display(df_largo)
```

```python
df_ancho = df_largo.pivot(index='Producto', columns='Mes', values='Ventas')
display(df_ancho)
```

```python
df_largo_de_nuevo = pd.melt(df_ancho.reset_index(), id_vars=['Producto'], value_vars=['Enero', 'Febrero'], var_name='Mes', value_name='Ventas')
display(df_largo_de_nuevo )
display(df_largo)
```
