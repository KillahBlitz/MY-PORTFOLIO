```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('sales_data.csv')
df['City'] = df['Purchase Address'].apply(lambda x: x.split(',')[1].strip())
df.head()
```

```python
product_counts = df['Product'].value_counts()
plt.figure(figsize=(4, 4))
plt.pie(product_counts, labels=product_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribución de Productos Vendidos')
plt.axis('equal')
plt.show()
```

```python
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6']
explode = [0.1] + [0] * (len(product_counts) - 1)

plt.figure(figsize=(4, 4))
plt.pie(product_counts, labels=product_counts.index, autopct='%1.1f%%', startangle=90, 
        colors=colors, explode=explode, shadow=True)
plt.title('Distribución de Productos Vendidos (Personalizado)')
plt.axis('equal')
plt.show()
```

```python
city_counts = df['City'].value_counts()
plt.figure(figsize=(4, 4))
plt.pie(city_counts, labels=city_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribución de Ventas por Ciudad')
plt.axis('equal')
plt.show()
```

```python
threshold = 8
city_counts['Others'] = city_counts[city_counts < threshold].sum()
city_counts = city_counts[city_counts >= threshold]
plt.figure(figsize=(4, 4))
plt.pie(city_counts, labels=city_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribución de Ventas por Ciudad (Agrupando Pequeñas)')
plt.axis('equal')
plt.show()
```
