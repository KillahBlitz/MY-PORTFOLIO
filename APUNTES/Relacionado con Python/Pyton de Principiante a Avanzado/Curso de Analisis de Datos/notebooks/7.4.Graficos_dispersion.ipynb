```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {'Height': [150, 160, 165, 170, 175, 180, 185, 190],
        'Weight': [50, 55, 60, 65, 70, 75, 80, 85]}
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(4, 3))
plt.scatter(df['Height'], df['Weight'], color='blue', marker='o')  # Dispersión con marcador circular
plt.title('Basic Scatter Plot - Matplotlib')  # Título del gráfico
plt.xlabel('Height (cm)')  # Etiqueta del eje X
plt.ylabel('Weight (kg)')  # Etiqueta del eje Y
plt.grid(True, linestyle='--', alpha=0.6)  # Cuadrícula ligera
plt.show()
```

```python
import seaborn as sns
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(4, 3))
sns.scatterplot(x='Height', y='Weight', data=df, color='green', s=50)
plt.title('Scatter Plot - Seaborn')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

```python
df['Category'] = ['Group 1', 'Group 1', 'Group 2', 'Group 2', 'Group 1', 'Group 2', 'Group 1', 'Group 2']
display(df)
```

```python
plt.figure(figsize=(4, 3))
sns.scatterplot(x='Height', y='Weight', hue='Category', data=df, palette='muted', s=50)
plt.title('Scatter Plot with Categories - Seaborn')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.legend(title='Category')
plt.show()
```

```python
plt.figure(figsize=(4, 3))
sns.regplot(x='Height', y='Weight', data=df, scatter_kws={'s':50}, line_kws={'color':'red'})
plt.title('Scatter Plot with Trend Line - Seaborn')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
df['Income'] = [80000, 32000, 35000, 40000, 45000, 48000, 52000, 60000]
display(df)
```

```python
plt.figure(figsize=(4, 3))
plt.scatter(df['Height'], df['Weight'], s=df['Income']*0.001, c='blue', alpha=0.6, edgecolor='w', linewidth=2)
plt.title('Bubble Chart')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
data_multi = {'Height': [150, 160, 165, 170, 175, 180, 185, 190],
              'Weight_A': [50, 55, 60, 65, 70, 75, 80, 85],
              'Weight_B': [48, 53, 58, 63, 68, 73, 78, 83]}
df_multi = pd.DataFrame(data_multi)
display(df_multi)
```

```python
plt.figure(figsize=(4, 3))
plt.scatter(df_multi['Height'], df_multi['Weight_A'], color='blue', s=40, label='Group A')
plt.scatter(df_multi['Height'], df_multi['Weight_B'], color='green', s=40, label='Group B')
plt.title('Scatter Plot with Multiple Series')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
