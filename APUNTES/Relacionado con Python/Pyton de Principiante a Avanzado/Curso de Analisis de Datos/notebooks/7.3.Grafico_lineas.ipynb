```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
        'Sales': [250, 300, 400, 350, 450, 500]}
df = pd.DataFrame(data)
display(df)
```

```python
plt.figure(figsize=(4, 3))
plt.plot(df['Month'], df['Sales'], color='blue', marker='o')
plt.title('Basic Line Chart - Matplotlib')  # Título del gráfico
plt.xlabel('Month')  # Etiqueta del eje X
plt.ylabel('Sales')  # Etiqueta del eje Y
plt.grid(True, linestyle='--', alpha=0.6)  # Agregar una cuadrícula
plt.show()
```

```python
import seaborn as sns
```

```python
sns.set_style("whitegrid") 

plt.figure(figsize=(4, 3))
sns.lineplot(x='Month', y='Sales', data=df, marker='o', color='red')
plt.title('Line Chart - Seaborn')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
data_multi = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
              'Product A': [250, 300, 400, 350, 450, 500],
              'Product B': [200, 250, 300, 280, 320, 400]}
df_multi = pd.DataFrame(data_multi)
display(df_multi)
```

```python
plt.figure(figsize=(4, 3))
plt.plot(df_multi['Month'], df_multi['Product A'], color='blue', marker='o', label='Product A')
plt.plot(df_multi['Month'], df_multi['Product B'], color='green', marker='o', label='Product B')
plt.title('Multi-line Chart')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
plt.figure(figsize=(4, 3))
plt.plot(df['Month'], df['Sales'], color='darkorange', marker='D', linestyle='--', linewidth=2, markersize=10)
plt.title('Customized Line Chart')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

- linestyle='--': Cambia el estilo de la línea a discontinua.
- marker='D': Define un marcador en forma de diamante.
- linewidth y markersize: Controlan el grosor de la línea y el tamaño de los marcadores.

```python
plt.figure(figsize=(4, 3))
sns.lineplot(x='Month', y='Sales', data=df, marker='o', errorbar='sd', color='purple')
plt.title('Line Chart with Confidence Interval - Seaborn')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
df_long = pd.melt(df_multi, id_vars=['Month'], value_vars=['Product A', 'Product B'], 
                  var_name='Product', value_name='Sales')
display(df_long)
```

```python
plt.figure(figsize=(4, 3))
sns.lineplot(x='Month', y='Sales', hue='Product', data=df_long, marker='o', errorbar='sd')
plt.title('Line Chart with Confidence Interval - Seaborn')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
import numpy as np
# Crear un DataFrame con múltiples observaciones por cada mes y producto
np.random.seed(42)  # Para reproducibilidad
data_multi_random = {'Month': ['Jan', 'Jan', 'Jan', 'Feb', 'Feb', 'Feb', 'Mar', 'Mar', 'Mar', 
                               'Apr', 'Apr', 'Apr', 'May', 'May', 'May', 'Jun', 'Jun', 'Jun'],
                     'Product': ['Product A', 'Product A', 'Product A', 'Product A', 'Product A', 'Product A', 
                                 'Product A', 'Product A', 'Product A', 'Product A', 'Product A', 'Product A', 
                                 'Product A', 'Product A', 'Product A', 'Product A', 'Product A', 'Product A'],
                     'Sales': [250, 260, 240, 300, 310, 290, 400, 410, 390, 350, 360, 340, 450, 460, 440, 500, 510, 490]}

df_multi_random = pd.DataFrame(data_multi_random)

display(df_multi_random)

# Añadir una pequeña variabilidad a las ventas de 'Product B'
data_multi_random_b = {'Month': ['Jan', 'Jan', 'Jan', 'Feb', 'Feb', 'Feb', 'Mar', 'Mar', 'Mar', 
                               'Apr', 'Apr', 'Apr', 'May', 'May', 'May', 'Jun', 'Jun', 'Jun'],
                     'Product': ['Product B', 'Product B', 'Product B', 'Product B', 'Product B', 'Product B', 
                                 'Product B', 'Product B', 'Product B', 'Product B', 'Product B', 'Product B', 
                                 'Product B', 'Product B', 'Product B', 'Product B', 'Product B', 'Product B'],
                     'Sales': [200, 210, 190, 250, 260, 240, 300, 310, 290, 280, 290, 270, 320, 330, 310, 400, 410, 390]}

df_multi_random_b = pd.DataFrame(data_multi_random_b)

# Concatenar ambos productos
df_random = pd.concat([df_multi_random, df_multi_random_b])

# Graficar con Seaborn mostrando el intervalo de confianza
plt.figure(figsize=(4, 3))
sns.lineplot(x='Month', y='Sales', hue='Product', data=df_random, marker='o', errorbar='sd')
plt.title('Line Chart with Confidence Interval - Seaborn')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```
