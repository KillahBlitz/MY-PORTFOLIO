```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#f(x)=x^2-4x+4
x = np.linspace(-1,5,400) #domino de (a,b) y el ultimo parametro es la cantidad de puntos
y = x**2 - 4*x + 4 #declaro la funcion

#funciones para graficar
plt.figure(figsize=(8,4)) #relacion de aspecto o tamanio de la grafica
plt.plot(x,y, label="f(x)=x^2-4x+4", color="red") #los parametros q voy a graficar, (x,y) texto de funcion y el color
plt.title("Grafico de la funcion cuadratica") #Colocar titulo
#colocar texto en ejes
plt.xlabel("x")
plt.ylabel("f(x)")

plt.grid(True) #muestra una cuadricula
plt.legend() #mostrar legenda
plt.show() #mostrar grafico
```

```python
#cambio de color
plt.plot(x,y, color="blue")
#colores con nombres en ingles: blue, red, yellow, brown, green, etc
#colores con codigo hexadecimal: #FFFFFF, #FF5733, etc
#Colores con codigo RGB: 1,1,1; 0.1,0.2,0.5; etc
plt.show()
```

```python
#cambiar el estilo de la linea
plt.plot(x,y, color=(0.9,0.2,0.5), linestyle="--")
#utlilizo formato rgb
#- es linea continua, -- es linea discontinua, -. es linea discontinua en intervalos grandes y chicos
#: son lineas punteadas
plt.show()
```

```python
#marcadores
#cambio de funcion
x=[4,6,7]
y=[3,5,2]
#cambio de color
plt.plot(x,y, color="#FF5733", linestyle="-.",marker="o")
#utlizo ahora formato hexadecimal
#linea de intervalos
#marcador de puntos, o son puntos, s son cuadrados, ^ triangulos, etc
plt.show()
```
