```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50]
}

df = pd.DataFrame(data)
display(df)
```

```python
df['sqrt_A'] = np.sqrt(df['A'])
display(df)
```

```python
df['cumsum_B'] = np.cumsum(df['B'])
display(df)
```

```python
data = {
    'x': np.random.rand(100),
    'y': np.random.rand(100)
}

df = pd.DataFrame(data)
display(df)
```

```python
#correlacion de pearson
correlacion, p_valor = stats.pearsonr(df['x'], df['y'])
print(f"correlacion de pearson {correlacion}, Pvalor {p_valor}")
```

```python
#regresion lineal simple 
slope, intercept, r_value, p_value, std_err = stats.linregress(df['x'], df['y'])
print(f"valores retornados:\n y = {slope} * x + {intercept} and {r_value}, {p_value}, {std_err}")
```

```python
#funciones universales de numpy
data = {
    'Valores': np.random.randn(100)
}

df = pd.DataFrame(data)
df
```

```python
#funcion exponencial a la columna de valores
df['exp_valores'] = np.exp(df['Valores'])
display(df)
```

```python
#funciones complejas con scipy
#creacion de muestra de datos
muestra = np.random.randn(100) + 2
```

```python
t_stat, p_Valor = stats.ttest_lsamp(muestra, 2)
print(f"Prueba t de una muestra: t-statistic = {t_stat}, p-valor = {p_valor}")
```
