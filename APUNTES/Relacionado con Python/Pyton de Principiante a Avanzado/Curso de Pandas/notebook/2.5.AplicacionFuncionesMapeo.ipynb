```python
import pandas as pd
```

```python
#Seleccion de fila o columna aplicando el metodo deseado
data = {
    "Producto": ["Manzanas","Naranjas","Platanos","Uvas","Peras"],
    "Precio": [100,80,60,120,90]
}

df = pd.DataFrame(data)
df
```

```python
#construir precios con descuentos
def descuento(x):
    return x*0.85
```

```python
df["Precio_descuento"] = df['Precio'].apply(descuento)
df
```

```python
#aplicacion con series
#funcion que convierta las letras a mayusculas
#ocupamos el metodo upper
cadena = "hola"
cadena.upper()

#uso de funciones anonimas (lambda)
df["Producto"] = df["Producto"].apply(lambda x: x.upper())
df
```

```python
#funciones map y applymap
#mapeo de valores
#regreso df a su valor inicial
df["Producto"] = df["Producto"].map({"Manzanas": "Apples", "Naranjas": "Oranges"})
df
#sustitucion de valores en ingles, sin embargo los que no estan en el diccionario se sustituye por NaN
```

```python
#funcion Applymap
#este pasa por todo el dataframe, todo lo que tenga que ver con string en este caso particular
df = df.applymap(lambda x: x.lower() if type(x) == str else x)
df
```

```python
#funcionalidad de apply mas compleja
df["Descuento"] = [10,15,10,20,5]
#agregacion de precio final
df["PrecioFinal"] = df.apply(lambda row: row["Precio"]*((100-row["Descuento"])/100), axis=1)
df
```

```python
#evitar el NaN en la funcion map
data = {
    "Producto": ["Manzanas","Naranjas","Platanos","Uvas","Peras"],
    "Precio": [100,80,60,120,90]
}

df = pd.DataFrame(data)
mapeo = {"Manzanas": "Apples", "Naranjas": "Oranges"}
df["Producto"] = df["Producto"].map(lambda x: mapeo.get(x,x))
df
```
