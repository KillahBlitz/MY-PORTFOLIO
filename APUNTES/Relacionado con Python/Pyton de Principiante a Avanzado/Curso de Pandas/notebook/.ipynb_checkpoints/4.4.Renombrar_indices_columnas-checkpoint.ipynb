```python
import pandas as pd
```

```python
data = {
    'Empleado': ['Juan','Ana','Pedro','Luis','Maria'],
    'Salario': [50000, 60000, 55000, 58000, 62000],
    'Anios_Experiencia': [5,7,6,8,9]
}

df = pd.DataFrame(data)
df
```

```python
#renombre de columnas
df_renombrado = df.rename(columns={'Salario': 'Salario_Anual', 'Anios_Experencia': 'Experiencia'})
df_renombrado
```

```python
#consultar columnas
df.columns = ['Empleado', 'Salario_Anual', 'Experiencia']
df
```

```python
#Indice de columnas
df_renombrado_indices = df.rename(index={0: 'Emp1', 1: 'Emp2', 2: 'Emp3', 3: 'Emp4', 4: 'Emp5'})
df_renombrado_indices
```

```python
df.index = ['Emp1','Emp2','Emp3','Emp4','Emp5']
df
```

```python
df_reorganizado = df.reindex(columns={'Experiencia', 'Salario_Anual', 'Empleado'})
df_reorganizado
df = df[['Experiencia', 'Salario_Anual', 'Empleado']]
df
```

```python
#aqui podemos ver como podemos mover los indices como querramos
df_reorganizado_filas = df.reindex(index=['Emp3','Emp1','Emp5','Emp4','Emp2'])
df_reorganizado_filas
#NOTA: columns es de columnas e index es de filas
```

```python
df_ordenado = df.sort_index()
df_ordenado
#ordenacion de los datos por indices
```
