```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#generacion de puntos aleatorios en eje x, y
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
#graficar estas funciones, que en este caso son puntos aleatorios
plt.scatter(x, y)
#Colocar titulo
plt.title("Grafico de dispersion basico")
#colocar texto en ejes
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.show()
```

```python
#Grafica de funciones
#f(x)=x^2-4x+4
#primero hay que generar puntos del dominio
x = np.linspace(-1,5,400) #domino de (a,b) y el ultimo parametro es la cantidad de puntos
#declaro la funcion
y = x**2 - 4*x + 4

#funciones para graficar
plt.figure(figsize=(8,4)) #relacion de aspecto o tamanio de la grafica
plt.plot(x,y, label="f(x)=x^2-4x+4", color="red") #los parametros q voy a graficar, (x,y) texto de funcion y el color
plt.title("Grafico de la funcion cuadratica") #Colocar titulo
#colocar texto en ejes
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True) #muestra una cuadricula
plt.legend() #mostrar legenda
plt.show() #mostrar grafico
```
