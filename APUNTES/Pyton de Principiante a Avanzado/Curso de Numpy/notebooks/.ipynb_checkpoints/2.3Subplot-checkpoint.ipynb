```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#uso de subplot
#declaracion de la variable independiente
x = np.linspace(0,10,100)
#declaracion de funciones
y_linear = x
y_exponential = np.exp(x)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(x,y_linear, "r-")
plt.title("Grafico de la funcion lineal") #Colocar titulo
#colocar texto en ejes
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.subplot(1,2,2) 
plt.plot(x,y_exponential, "b-")
plt.title("Grafico de la funcion exponencial") #Colocar titulo
#colocar texto en ejes
plt.xlabel("x")
plt.ylabel("g(x)")
plt.grid(True)
plt.show()
#en conclusion, el metodo subplot es un subindice para graficar funciones paralelamente o dentro de la misma grafica
```

```python
#graficando desde la misma grafica
plt.figure(figsize=(8,4))
plt.plot(x,y_linear,label="f(x)=x", color="yellow", linestyle="-")
plt.plot(x,y_exponential,label="g(x)=e^x", color="green", linestyle="--")
plt.title("Grafico de las funciones f(x) y g(x)") #Colocar titulo
#colocar texto en ejes
plt.xlabel("x")
plt.ylabel("f(x) and g(x)")
plt.ylim(0,10)
plt.grid(True)
plt.show()
```

```python
#metodo subplots
#crea una figura y conjunto de graficos en una sola llamada
#Crear figura y arreglo de ejes (2x2)
fig, axs = plt.subplots(2,2,figsize=(10,10)) #2 filas, 2 columnas

#Variable independiente
x = np.linspace(0, 2*np.pi,100)
funcs = [np.sin(x),np.cos(x),np.tan(x),np.sqrt(x)]

#Titulos de cada grafica
titles = ["seno","coseno","tangente","Raiz Cuadrada"]

#llenar cada subgrafico con una funcion
for ax, func, title in zip(axs.ravel(), funcs, titles):
    ax.plot(x,func)
    ax.set_title(title)
    ax.grid(True)

#ajustar automaticamente el layout de los subgraficos para evitar solapamientos
plt.tight_layout()
plt.show()
```
