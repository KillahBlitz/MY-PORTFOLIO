```python
import pandas as pd
```

```python
data = {
    'producto': ['Manzana', 'Manzana', 'Banana', 'Banana', 'Cereza', 'Cereza',  'Manzana', 'Banana', 'Cereza'],
    'categoria': ['Fruta', 'Fruta', 'Fruta', 'Fruta', 'Fruta', 'Fruta', 'Fruta', 'Fruta', 'Fruta'],
    'ventas': [50, 30, 20, 15, 10, 25, 45, 55, 40],
    'precio': [0.5, 0.5, 0.3, 0.3, 0.8, 0.8, 0.5, 0.3, 0.8],
    'fecha': pd.date_range(start='2024-01-01', periods=9, freq='D')
}

df = pd.DataFrame(data)
display(df)
```

```python
#funciones personalizadas con el metodo apply
#definir funcion
def calcularingresototal(row):
    return row['ventas']*row['precio']

df['preciototal'] = df.apply(calcularingresototal, axis=1)
display(df)
```

```python
#funcion que clasifica los productos dependiendo de las ventas
def clasificar(ventas):
    if ventas > 40:
        return 'Alto'
    elif 30 <= ventas <= 40:
        return 'Medio'
    else:
        return 'Bajo'

df['clasificacion'] = df['ventas'].apply(clasificar)
display(df)
```

```python
#transformaciones de texto
#pasar texto a mayuscula
df['producto_upper'] = df['producto'].apply(lambda x: x.upper())
display(df)
```

```python
#aplicar funciones a varias columnas
def calcularporc(ventas):
    total = ventas.sum()
    return ventas / total * 100

df['porcentajeventas'] = df.groupby('producto')['ventas'].transform(calcularporc)
display(df)
```
