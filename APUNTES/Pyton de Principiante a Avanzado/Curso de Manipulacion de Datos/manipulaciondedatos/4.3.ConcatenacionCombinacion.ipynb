```python
import pandas as pd
```

```python
data_lima = {
    'Producto': ['A','B'],
    'Ventas': [250,150],
    'Ciudad': ['Lima','Lima']
}

df_lima = pd.DataFrame(data_lima)

data_bogota = {
    'Producto': ['A','C'],
    'Ventas': [100,200],
    'Ciudad': ['Bogota','Bogota']
}

df_bogota = pd.DataFrame(data_bogota)

display(df_lima)
display(df_bogota)
```

```python
#concatenacion de manera vertical
df_concatenado = pd.concat([df_lima, df_bogota]).reset_index(drop=True)
display(df_concatenado)
```

```python
data_inventario = {
    'Producto': ['A','B','C'],
    'Inventario': [30,45,25]
}

df_inventario = pd.DataFrame(data_inventario)
display(df_inventario)
```

```python
#concatenacion de manera horizontal
df_concatenado_h = pd.concat([df_concatenado, df_inventario], axis=1)
display(df_concatenado_h)
#en este caso no es lo optimo concatenar estos dataframes
```

```python
#metodo merge
df_combinado = pd.merge(df_concatenado, df_inventario, on='Producto', how='left')
display(df_combinado)
```

```python
#tipos de combinaciones con merge
#innerjoin
df_combinado = pd.merge(df_concatenado, df_inventario, on='Producto', how='inner')
display(df_combinado)
```

```python
#rightjoin
df_combinado = pd.merge(df_concatenado, df_inventario, on='Producto', how='right')
display(df_combinado)
```

```python
#outerjoin
df_combinado = pd.merge(df_concatenado, df_inventario, on='Producto', how='outer')
display(df_combinado)
```
