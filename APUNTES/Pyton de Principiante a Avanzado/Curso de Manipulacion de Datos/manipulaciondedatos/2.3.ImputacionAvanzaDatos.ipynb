```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from fancyimpute import IterativeImputer
```

```python
df= pd.read_csv('datos_con_faltantes.csv')
display(df)
```

```python
#Tecnicas Avanzadas de Imputacion de Datos

#Texto se rellana con texto
df['Nombre'] = df['Nombre'].fillna('Desconocido')
df['Ciudad'] = df['Ciudad'].fillna('No registra')

#para datos numericos es conveninete rellenar con datos adecuados, como medidas de tendencia central
df['Salario'] = df['Salario'].fillna(df['Salario'].mean())
df['Edad'] = df['Edad'].fillna(df['Edad'].median())
display(df)
#NOTA: se puede usar la moda pero recordar que hay valores bimodales y trimodales, recordar sintaxis mode()[n]
```

```python
#tecnicas avanzadas de Imputacion
df= pd.read_csv('datos_con_faltantes.csv')
display(df)

df['Nombre'] = df['Nombre'].fillna('Desconocido')
df['Ciudad'] = df['Ciudad'].fillna(df['Ciudad'].mode()[0])
#Uso de Imputacion por regresion (uso de libreria sklearn)
#Regresion Lineal
model = LinearRegression()
model.fit(df[['Edad']].dropna(), df['Salario'].dropna())
#uso de apply para llenar los datos
df['Salario'] = df.apply(lambda row: model.predict([[row['Edad']]])[0] if pd.isnull(row['Salario']) else row['Salario'], axis=1)

display(df)
```
