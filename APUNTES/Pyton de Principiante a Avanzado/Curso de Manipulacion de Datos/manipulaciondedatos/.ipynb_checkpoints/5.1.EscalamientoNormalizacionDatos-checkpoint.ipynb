```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
data = {
    'producto': ['Manzana', 'Banana', 'Cereza', 'Durazno', 'Pera'],
    'unitario': [0.5,0.3,0.8,1.2,0.9],
    'ventas': [50,30,70,85,40],
    'descuento': [5,2,7,8,5]
}

df = pd.DataFrame(data)
display(df)
```

```python
#ESCALAMIENTO por minima scaling
scaler = MinMaxScaler()
columnasNumericas = ['unitario', 'ventas','descuento']
#para buena practica crear una copia del data frame original
df_escalado = df.copy()

df_escalado[columnasNumericas] = scaler.fit_transform(df[columnasNumericas])
display(df_escalado)
#basicamente hace un escalado de 0 a 1 donde0 es el valor mas bajo y 1 es el valor mas alto
```

```python
#Normalizacion de datos con standar scaler
scalernormalizacion = StandardScaler()
df_normalizado = df.copy()
df_normalizado[columnasNumericas] = scalernormalizacion.fit_transform(df[columnasNumericas])
display(df_normalizado)
```

```python
#visualizacion de los datos
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
#diagrama de cajas
df[columnasNumericas].plot(kind='box', title="Original", ax=axes[0])
df_escalado[columnasNumericas].plot(kind='box', title="Escalado", ax=axes[1])
plt.show()
```
