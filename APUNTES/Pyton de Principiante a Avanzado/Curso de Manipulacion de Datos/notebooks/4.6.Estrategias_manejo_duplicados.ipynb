```python
import pandas as pd

data = {
    'ID': [1, 2, 3, 4, 5, 3, 7, 8, 9, 10, 2],
    'Nombre': ['Ana', 'Luis', 'Carlos', 'María', 'Elena', 'Carlos', 'Sofía', 'Andrés', 'Laura', 'Ana', 'Luis'],
    'Edad': [23, 35, 30, 22, 28, 30, 34, 45, 29, 41, 35],
    'Salario': [50000, 60000, 58000, 58000, 49000, 58000, 72000, 61000, 52000, 67000, 60000],
    'Fecha_Ingreso': ['2020-01-10', '2019-06-15', '2020-03-25', '2018-07-30', '2018-07-30', 
                      '2020-03-25', '2016-11-22', '2021-04-01', '2015-05-19', '2014-08-25', '2019-06-15']
}

df_duplicated = pd.DataFrame(data)
display(df_duplicated)
```

```python
df_no_duplicates = df_duplicated.drop_duplicates(subset=["Nombre"],keep='last')
display(df_no_duplicates)
```

```python
df_summarized = df_duplicated.groupby('Nombre').agg({
    'Edad': 'first', 
    'Salario': 'mean',
    'Fecha_Ingreso': 'first'
}).reset_index()

display(df_summarized)
```

```python
df_duplicated['Es_Duplicado'] = df_duplicated.duplicated(subset=["Nombre"])
display(df_duplicated)
```

```python
df_nuevo = pd.DataFrame({
    'ID': [1, 2, 11],
    'Nombre': ['Ana', 'Luis', 'Gabriela'],
    'Edad': [24, 36, 27],
    'Salario': [51000, 61000, 68000],
    'Fecha_Ingreso': ['2021-01-15', '2021-06-15', '2021-08-01']
})
```

```python
df_merged = pd.concat([df_duplicated, df_nuevo]).drop_duplicates(subset=['ID'], keep='last')
display(df_merged)
```
