```python
import pandas as pd

# Crear un DataFrame de ejemplo
data = {
    'producto': ['Manzana', 'Manzana', 'Banana', 'Banana', 'Cereza', 'Cereza', 'Manzana', 'Banana', 'Cereza'],
    'categoría': ['Fruta', 'Fruta', 'Fruta', 'Fruta', 'Fruta', 'Fruta', 'Fruta', 'Fruta', 'Fruta'],
    'ventas': [50, 30, 20, 15, 10, 25, 45, 55, 40],
    'precio_unitario': [0.5, 0.5, 0.3, 0.3, 0.8, 0.8, 0.5, 0.3, 0.8],
    'fecha': pd.date_range(start='2024-01-01', periods=9, freq='D')
}

df = pd.DataFrame(data)
display(df)
```

```python
resumen_ventas_por_producto = df.groupby('producto')['ventas'].agg(['sum', 'mean']).reset_index()
display(resumen_ventas_por_producto)
```

```python
def rango_ventas(grupo):
    return grupo.max() - grupo.min()

rango_por_producto = df.groupby('producto')['ventas'].agg(rango_ventas).reset_index()
display(rango_por_producto)
```

```python
df['ventas_normalizadas'] = df.groupby('producto')['ventas'].transform(lambda x: (x - x.mean()) / x.std())
display(df[['producto', 'ventas', 'ventas_normalizadas']])
```

```python
df['ingreso_total'] = df['ventas'] * df['precio_unitario']
df_con_ingreso_total = df.groupby(['producto', 'categoría']).agg({
    'ventas': 'sum',
    'ingreso_total': 'sum'
}).reset_index()
display(df_con_ingreso_total)
```
