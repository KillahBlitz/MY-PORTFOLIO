```python
import pandas as pd
```

```python
df = pd.read_csv('datos_con_faltantes.csv')
display(df)
```

```python
df['Nombre'] = df['Nombre'].fillna('Desconocido')
display(df)
```

```python
df['Ciudad'] = df['Ciudad'].fillna('No registra')
display(df)
```

```python
df['Salario'] = df['Salario'].fillna(df['Salario'].mean())
display(df)
```

```python
df['Edad'] = df['Edad'].fillna(df['Edad'].median())
display(df)
```
