```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.rand(10, 12) 
print(data)
```

```python
plt.figure(figsize=(5, 4))
sns.heatmap(data, cmap='Greens')
plt.title('Basic Heatmap - Greens')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data, annot=True, fmt=".1f", cmap='coolwarm')
plt.title('Heatmap with Annotations - Seaborn')
plt.show()
```

```python
import pandas as pd
data = {'A': [0, 2, 3, 4, 5], 
        'B': [5, 8, 3, 2, 1], 
        'C': [2, 3, 4, 5, 6], 
        'D': [5, 3, 1, 4, 2]}
df = pd.DataFrame(data)
display(df)
```

```python
corr_matrix = df.corr()
display(corr_matrix)
```

```python
plt.figure(figsize=(4, 3))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
plt.figure(figsize=(4, 3))
sns.heatmap(df, vmin=0, vmax=10, cmap='YlGnBu', cbar_kws={'label': 'Value'})
plt.title('Heatmap with Color Limits')
plt.show()
```

```python
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

plt.figure(figsize=(4, 3))
sns.heatmap(corr_matrix, mask=mask, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap with Color Limits')
plt.show()
```

```python
plt.figure(figsize=(4, 3))
sns.heatmap(df, annot=True, cmap='magma', linecolor='white', linewidths=1)
plt.title('Customized Heatmap - Seaborn')
plt.xticks(rotation=45)
plt.show()
```
