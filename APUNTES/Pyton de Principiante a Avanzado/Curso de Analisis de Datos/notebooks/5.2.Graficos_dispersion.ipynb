```python
import pandas as pd
df = pd.read_csv('sales_data_correlated.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5, 3))
plt.scatter(df['Price Each'], df['Quantity Ordered'], alpha=0.5)
plt.title('Gráfico de Dispersión: Price Each vs. Quantity Ordered')
plt.xlabel('Price Each')
plt.ylabel('Quantity Ordered')
plt.show()
```

```python
import seaborn as sns
plt.figure(figsize=(5, 3))
sns.scatterplot(data=df, x='Price Each', y='Quantity Ordered')
plt.title('Gráfico de Dispersión: Price Each vs. Quantity Ordered')
plt.xlabel('Price Each')
plt.ylabel('Quantity Ordered')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x='Price Each', y='Quantity Ordered', hue='Product', palette='viridis')
plt.title('Gráfico de Dispersión: Price Each vs. Quantity Ordered (Coloreado por Producto)')
plt.xlabel('Price Each')
plt.ylabel('Quantity Ordered')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.regplot(data=df, x='Price Each', y='Quantity Ordered', scatter_kws={'alpha':0.5})
plt.title('Gráfico de Dispersión con Línea de Tendencia')
plt.xlabel('Price Each')
plt.ylabel('Quantity Ordered')
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.scatterplot(data=df, x='Price Each', y='Quantity Ordered', s=40, alpha=0.6, edgecolor='k')
plt.title('Gráfico de Dispersión Personalizado')
plt.xlabel('Price Each')
plt.ylabel('Quantity Ordered')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
