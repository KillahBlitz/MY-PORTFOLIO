```python
import pandas as pd
df = pd.read_csv('sales_data.csv')
df.head()
```

```python
df['City'] = df['Purchase Address'].apply(lambda x: x.split(',')[1].strip())
df.head()
```

```python
product_frequency = df['Product'].value_counts()
print(product_frequency)
```

```python
city_frequency = df['City'].value_counts()
print(city_frequency)
```

```python
product_proportion = df['Product'].value_counts(normalize=True)
print(product_proportion)
```

```python
city_proportion = df['City'].value_counts(normalize=True)
print(city_proportion)
```

```python
product_city_crosstab = pd.crosstab(df['City'], df['Product'])
display(product_city_crosstab)
```

```python
product_by_city = df.groupby('City')['Product'].value_counts()
print(product_by_city)
```

```python
high_price_products = df[df['Price Each'] > 1000]
high_price_proportion = high_price_products['City'].value_counts(normalize=True)
print(high_price_proportion)
```
