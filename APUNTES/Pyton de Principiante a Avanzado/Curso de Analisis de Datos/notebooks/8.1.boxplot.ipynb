```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {'Category A': [25, 30, 35, 45, 50, 55, 60, 70, 80, 85],
        'Category B': [20, 25, 30, 35, 40, 45, 50, 55, 65, 70]}
df = pd.DataFrame(data)
display(df)
```

```python
plt.figure(figsize=(4, 3))
plt.boxplot([df['Category A'], df['Category B']], tick_labels=['Category A', 'Category B'])
plt.title('Boxplot - Matplotlib')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

data = {'Category': ['A']*10 + ['B']*10,
        'Values': [25, 30, 35, 45, 50, 55, 60, 70, 80, 85, 20, 25, 30, 35, 40, 45, 50, 55, 65, 70]}
df_seaborn = pd.DataFrame(data)
display(df_seaborn)
```

```python
plt.figure(figsize=(4, 3))
sns.boxplot(x='Category', y='Values', hue='Category' ,data=df_seaborn, palette='pastel')
plt.title('Boxplot - Seaborn')
plt.show()
```

```python
plt.figure(figsize=(4, 3))
sns.boxplot(x='Category', y='Values', hue='Category' ,data=df_seaborn, palette='pastel')
sns.swarmplot(x='Category', y='Values', data=df_seaborn, color='red', size=7)
plt.title('Boxplot - Seaborn')
plt.show()
```

```python
data_subcat = {'Category': ['A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B'],
               'Group': ['X', 'X', 'Y', 'Y', 'Y', 'X', 'X', 'Y', 'Y', 'Y'],
               'Values': [25, 30, 35, 45, 50, 20, 25, 30, 35, 40]}
df_subcat = pd.DataFrame(data_subcat)
display(df_subcat)
```

```python
plt.figure(figsize=(4, 3))
sns.boxplot(x='Category', y='Values', hue='Group', data=df_subcat, palette='Set2')
plt.title('Boxplot with Subcategories - Seaborn')
plt.show()
```
