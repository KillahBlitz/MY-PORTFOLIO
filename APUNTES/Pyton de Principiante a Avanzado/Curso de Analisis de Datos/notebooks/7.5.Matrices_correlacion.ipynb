```python
import pandas as pd
import numpy as np
```

- 1 indica una correlación positiva perfecta.
- -1 indica una correlación negativa perfecta.
- 0 indica que no hay correlación entre las variables.

```python
data = {'Height': [150, 160, 165, 170, 175, 180, 185, 190],
        'Weight': [50, 55, 60, 65, 70, 75, 80, 85],
        'Income': [30000, 32000, 35000, 40000, 45000, 48000, 52000, 60000]}
df = pd.DataFrame(data)
correlation_matrix = df.corr()
display(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(4, 3))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
strong_corr = correlation_matrix[(correlation_matrix > 0.98) | (correlation_matrix < -0.98)]
display(strong_corr)
```

```python
plt.figure(figsize=(4, 3))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', linecolor='white', linewidths=2)
plt.title('Customized Correlation Matrix Heatmap')
plt.xticks(rotation=45)
plt.show()
```

```python
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

plt.figure(figsize=(4, 3))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', mask=mask, linewidths=0.5)
plt.title('Heatmap with Mask')
plt.show()
```
