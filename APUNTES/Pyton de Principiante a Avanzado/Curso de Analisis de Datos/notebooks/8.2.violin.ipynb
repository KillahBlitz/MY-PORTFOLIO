```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {'Category': ['A']*10 + ['B']*10,
        'Values': [25, 30, 35, 45, 50, 55, 60, 70, 80, 85, 20, 25, 30, 35, 40, 45, 50, 55, 65, 70]}
df = pd.DataFrame(data)
display(df)
```

```python
plt.figure(figsize=(4, 3))
sns.violinplot(x='Category', y='Values',hue= 'Category',data=df, palette='pastel')
plt.title('Violin Plot - Seaborn')
plt.show()
```

```python
data_split = {'Category': ['A', 'A', 'A', 'A', 'A', 'B', 'B', 'B', 'B', 'B'],
              'Group': ['X', 'X', 'Y', 'Y', 'Y', 'X', 'X', 'Y', 'Y', 'Y'],
              'Values': [25, 30, 35, 45, 50, 20, 25, 30, 35, 40]}
df_split = pd.DataFrame(data_split)
display(df_split)
```

```python
plt.figure(figsize=(4, 3))
sns.violinplot(x='Category', y='Values', hue='Group', data=df_split, split=True, palette='Set2')
plt.title('Split Violin Plot - Seaborn')
plt.show()
```

```python
plt.figure(figsize=(4, 3))
sns.violinplot(x='Category', y='Values',hue='Category', data=df, palette='muted')
sns.boxplot(x='Category', y='Values',hue='Category', data=df, width=0.1, palette='dark:black', showcaps=True, boxprops={'facecolor':'None'})
plt.title('Violin Plot with Boxplot - Seaborn')
plt.show()
```

```python
plt.figure(figsize=(4, 3))
sns.violinplot(x='Values', y='Category',hue='Category', data=df, palette='pastel')
plt.title('Horizontal Violin Plot - Seaborn')
plt.show()
```

```python
plt.figure(figsize=(4, 3))
sns.violinplot(x='Category', y='Values', hue='Category', data=df, inner='point', palette='Set2')
plt.title('Violin Plot without Inner Lines - Seaborn')
plt.show()
```
