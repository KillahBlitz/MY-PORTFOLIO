```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {'Category': ['A', 'B', 'C', 'D'], 'Values': [25, 40, 35, 30]}
df = pd.DataFrame(data)
display(df)
```

```python
plt.figure(figsize=(4, 3)) # Establecer el tamaño del gráfico
plt.bar(df['Category'], df['Values'], color='lightblue') # Gráfico de barras con color
plt.title('Basic Bar Chart - Matplotlib')  # Título del gráfico
plt.xlabel('Category')  # Etiqueta del eje X
plt.ylabel('Values')  # Etiqueta del eje Y
plt.grid(True, axis='y', linestyle='--', alpha=0.6)
plt.show()
```

- plt.bar(x, height, color=) permite establecer el color de las barras.
- figsize=(width, height) define el tamaño del gráfico.
- plt.grid() agrega una cuadrícula, útil para mejorar la legibilidad del gráfico.

```python
import seaborn as sns
```

```python
sns.set_style("whitegrid")  # Estilo de fondo con cuadrícula
sns.set_palette("pastel")  # Paleta de colores

plt.figure(figsize=(4, 3))
sns.barplot(x='Category', y='Values', data=df)
plt.title('Bar Chart - Seaborn')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

- hue: Si tienes más categorías, puedes dividir las barras por color con la opción hue, ideal para comparar varias series de datos.
- ci: Por defecto, Seaborn agrega intervalos de confianza (barra de error). Puedes desactivarlo con ci=None.

```python
plt.figure(figsize=(4, 3))
plt.barh(df['Category'], df['Values'], color='lightgreen') 
plt.title('Horizontal Bar Chart')
plt.xlabel('Values')
plt.ylabel('Category')
plt.grid(True, axis='x', linestyle='--', alpha=0.6)
plt.show()
```

- plt.barh(x, y) se usa para crear un gráfico de barras horizontal.
- Este formato es útil para evitar superposiciones en las etiquetas largas.

```python
data_stacked = {'Category': ['A', 'B', 'C', 'D'],
                'Group 1': [25, 35, 30, 20],
                'Group 2': [15, 25, 20, 25]}
df_stacked = pd.DataFrame(data_stacked)
display(df_stacked )
```

```python
plt.figure(figsize=(4, 3))
plt.bar(df_stacked['Category'], df_stacked['Group 1'], label='Group 1', color='lightblue')
plt.bar(df_stacked['Category'], df_stacked['Group 2'], bottom=df_stacked['Group 1'], label='Group 2', color='lightcoral')
plt.title('Stacked Bar Chart')
plt.xlabel('Category')
plt.ylabel('Values')
plt.legend()
plt.grid(True, axis='y', linestyle='--', alpha=0.6)
plt.show()
```

- bottom: Este argumento coloca una barra encima de otra, creando el efecto apilado.
- label: Para definir una leyenda, útil para identificar las partes en el gráfico apilado.

```python
plt.figure(figsize=(4, 3))
width = 0.3 

categories = df_stacked['Category']
index = range(len(categories))


plt.bar([i - width/2 for i in index], df_stacked['Group 1'], width=width, label='Group 1', color='lightblue')
plt.bar([i + width/2 for i in index], df_stacked['Group 2'], width=width, label='Group 2', color='lightcoral')

plt.title('Grouped Bar Chart')
plt.xlabel('Category')
plt.ylabel('Values')

plt.xticks(index, categories)
plt.legend()

plt.grid(True, axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
plt.figure(figsize=(4, 3))
plt.bar(df['Category'], df['Values'], color='cornflowerblue', edgecolor='black', linewidth=2, alpha=0.8)
plt.title('Customized Bar Chart')
plt.xlabel('Category')
plt.ylabel('Values')
plt.xticks(rotation=45, fontsize=12)
plt.yticks(rotation=45, fontsize=12)
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.show()
```
