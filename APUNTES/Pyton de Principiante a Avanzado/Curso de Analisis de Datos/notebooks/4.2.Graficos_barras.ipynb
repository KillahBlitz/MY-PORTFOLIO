```python
import pandas as pd
df = pd.read_csv('sales_data.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(5, 3))
sns.countplot(data=df, x='Product', order=df['Product'].value_counts().index)
plt.title('Frecuencia de Productos Vendidos')
plt.xlabel('Producto')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.countplot(data=df, y='Product', order=df['Product'].value_counts().index)
plt.title('Frecuencia de Productos Vendidos (Horizontal)')
plt.xlabel('Cantidad Vendida')
plt.ylabel('Producto')
plt.show()
```

```python
df['City'] = df['Purchase Address'].apply(lambda x: x.split(',')[1].strip())
df.head()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='City', hue='Product', order=df['City'].value_counts().index)
plt.title('Frecuencia de Productos Vendidos por Ciudad')
plt.xlabel('Ciudad')
plt.ylabel('Cantidad Vendida')
plt.xticks(rotation=45)
plt.legend(title='Producto', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
product_proportion = df['Product'].value_counts(normalize=True).reset_index()
product_proportion.columns = ['Product', 'Proportion']
display(product_proportion)
```

```python
plt.figure(figsize=(5, 3))
sns.barplot(data=product_proportion, x='Product', y='Proportion')
plt.title('Proporción de Productos Vendidos')
plt.xlabel('Producto')
plt.ylabel('Proporción')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(5, 3))
sns.barplot(data=product_proportion, x='Product', y='Proportion', palette='viridis')
plt.title('Proporción de Productos Vendidos (Personalizado)')
plt.xlabel('Producto')
plt.ylabel('Proporción')
plt.xticks(rotation=45)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()
```
