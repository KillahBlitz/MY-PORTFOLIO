```python
import pandas as pd
df = pd.read_csv('sales_data_correlated.csv')
df.head()
```

```python
correlation_matrix = df[['Order ID','Quantity Ordered','Price Each']].corr()
print(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriz de Correlación')
plt.show()
```

```python
correlation_value = df['Price Each'].corr(df['Quantity Ordered'])
print(f"Correlación entre Price Each y Quantity Ordered: {correlation_value}")
```

```python
strong_correlations = correlation_matrix[(correlation_matrix > 0.7) | (correlation_matrix < -0.7)]
print(strong_correlations)
```

```python
sns.pairplot(df[['Price Each', 'Quantity Ordered']])
plt.show()
```
