```python
import pandas as pd
```

```python
data = {
    "producto": ["Manzanas","Naranjas","Platanos","Uvas","Peras"],
    "Precio": [100,80,60,120,90]
}

df = pd.DataFrame(data)
df
```

```python
#reindexacion
df.set_index("producto", inplace=True)
df
```

```python
#definir nuevo orden
nuevo_orden = ["Uvas", "Manzanas", "Naranjas", "Peras", "Platanos", "Melones"] #se puede agregar un nuevo producto
df_reindexado = df.reindex(nuevo_orden)
df_reindexado
```

```python
#deshacer y cambiar nombres
df_renombrado=df.rename(index={"Manzanas": "Apples","Naranjas": "Oranges"})
df_renombrado
```

```python
#resetear los indices
df_reset = df.reset_index()
df_reset
```
