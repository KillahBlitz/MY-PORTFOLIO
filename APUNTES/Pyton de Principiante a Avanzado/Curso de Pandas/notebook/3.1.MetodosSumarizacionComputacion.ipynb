```python
import pandas as pd
```

```python
data = {
    "producto": ["Manzanas","Naranjas","Platanos","Uvas","Peras"],
    "Precio": [100,80,60,120,90],
    "stock": [30,50,20,60,40]
}

df = pd.DataFrame(data)
df
```

```python
#sumatoria de los precios
suma_precios = df["Precio"].sum()
print("Sumatoria de precios: ", suma_precios)
#media de precios
media_precios = df["Precio"].mean()
print("Media de precios: ", media_precios)
#mediana de precios
mediana_precios = df["Precio"].median()
print ("media de precios: ", mediana_precios)
#maximos o minimos de precios
min_precio = df["Precio"].min()
max_precio = df["Precio"].max()
print("Valor maximo: ", max_precio)
print("Valor Minimo: ", min_precio)
#desviacion estandar y varianza
std_precio = df["Precio"].std()
var_precio = df["Precio"].var()
print("desviacion estandar: ", std_precio)
print("varianza: ", var_precio)

#metodo describe que me dira varias stats de los datos
resumen_precios = df["Precio"].describe()
print("Aplicacion de metodos")
print(resumen_precios)
```

```python
#metodos de computacion
#metodo cumsum, suma acumulada 
cumsum_precios = df["Precio"].cumsum()
cumsum_precios
#metodo de cumprod, producto acumulado (multiplicacion)
cumprod_precios = df["Precio"].cumprod()
cumprod_precios
#diferencia discreta a lo largo de un eje
diff_precios = df["Precio"].diff()
diff_precios
#cambio porcentual respecto al eje
pct_change_precios = df["Precio"].pct_change()
pctstring = f"{pct_change_precios*100}%"
print(pctstring)

df["ProductoAcumulado"] = cumprod_precios
df["SumasAcumuladas"] = cumsum_precios
df["DiferenciaDiscreta"] = diff_precios
df["CambioPorcentual"] = pct_change_precios
df
```
