```python
import pandas as pd
```

```python
data = {
    'Empleado': ['Juan','Ana','Pedro','Luis','Maria','Ana','Pedro','Luis','Maria'],
    'Departamento': ['Ventas','Marketing','Ventas','IT','Marketing','IT','Ventas','Marketing','Ventas'],
    'Salario': [50000,60000,55000,58000,62000,60000,55000,58000,62000],
    'Experiencia': [5,7,6,8,9,7,6,8,9]
}

df = pd.DataFrame(data)
display(df)
```

```python
#normalizar salarios por departamento
df['Salario_Normalizado'] = df.groupby('Departamento')['Salario'].transform(lambda x: (x-x.mean())/x.std())
#a cada valor se le resta la media y se divide por la desviacion estandar, metodo para normalizar
display(df)
```

```python
#obviamente, filtrado de datos con condicionales
df_filtrado = df.groupby('Departamento').filter(lambda x: len(x)>2)
display(df_filtrado)
```

```python
#funcion .agg
#aplicar funciones diferentes
agregaciones_diferentes = df.groupby('Departamento').agg({
    'Salario': 'sum',
    'Experiencia': 'mean'
})
display(agregaciones_diferentes)
```

```python
#agrupacion por metodo grouper que es para agrupaciones temporales
data1 = {
    'Fecha': pd.date_range(start='2023-01-01', periods=90,freq='D'),
    'Ventas': [i*100 for i in range(90)]
}
df_ventas = pd.DataFrame(data1)
#indexar por fecha
df_ventas.set_index('Fecha', inplace=True)
display(df_ventas)
```

```python
#agrupacion por mes y ver las ventas mensuales
ventas_mensuales = df_ventas.groupby(pd.Grouper(freq='ME')).sum()
display(ventas_mensuales)
```

```python
#operacion personalizada por cada grupo
#creacion de funcion de rangos salarios
def rango_salarios(grupo):
    return grupo['Salario'].max() - grupo['Salario'].min()

rango_salarios_PD = df.groupby('Departamento').apply(rango_salarios)
display(rango_salarios_PD)
```
