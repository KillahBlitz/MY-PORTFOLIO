```python
import pandas as pd
```

```python
data = {
    'Empleado': ['Juan','Ana','Pedro','Luis','Maria', 'Elias'],
    'Salario': [50000, 60000, 55000, 58000, 120000, 500],
    'Anios_Experiencia': [5,7,6,8,9, 1]
}

df = pd.DataFrame(data)
df
```

```python
#formas de detectar datos atipicos
media_salario = df['Salario'].mean()
std_salario = df['Salario'].std()

#definicion de umbral para valores atipicos
#valores que se desvien de la media dos veces (dos desviaciones estandar)
umbral = 1
#este valor puede ser distinto dependiendo del contexto del programa y los datos

#aqui hago el condicional de la desviacion estandar
df['Es_Atipico'] = abs(df['Salario'] - media_salario) > umbral * std_salario
df
```

```python
#ver valores atipicos usando perceptiles o intercuartiles
Q1 = df['Salario'].quantile(0.25) #primer cuartil
Q3 = df['Salario'].quantile(0.75) #tercer cuartil
IQR = Q3 - Q1 #rango de cuartiles

#definicion de umbrales
umbral_inferior = Q1 - 1.5 * IQR
umbral_superior = Q3 + 1.5 * IQR 
#formulas de estadistica

df['Es_Atipico'] = (df['Salario'] < umbral_inferior) | (df['Salario'] > umbral_superior)
df
```

```python
#ya encontramos los datos atipicos, ahora vamos a filtrar los datos atipicos

#creamos un nuevo DF para filtrar los atipicos
df_sin_atipicos = df[~df['Es_Atipico']] #se niega los valores true y false
df_sin_atipicos
```

```python
df_sin_atipicos = df[df['Es_Atipico']] #caso inverso de ver casos atipicos
df_sin_atipicos
```

```python
#remplazo de valores atipicos
df.loc[df['Es_Atipico'], 'Salario'] = media_salario
df
```
