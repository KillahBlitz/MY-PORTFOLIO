```python
import pandas as pd
```

```python
data = {
    'Empleado': ['Juan','Ana','Pedro','Luis','Maria','Ana','Pedro','Luis','Maria'],
    'Departamento': ['Ventas','Marketing','Ventas','IT','Marketing','IT','Ventas','Marketing','Ventas'],
    'Salario': [50000,60000,55000,58000,62000,60000,55000,58000,62000],
    'Experiencia': [5,7,6,8,9,7,6,8,9]
}

df = pd.DataFrame(data)
display(df)
```

```python
#tabla dinamica que demuestre la suma de salarios por departamento
pivot_table = df.pivot_table(values='Salario', index='Departamento', aggfunc='mean')
display(pivot_table)
```

```python
#tabla dinamica con multiples funciones
pivot_table = df.pivot_table(values='Salario', index='Departamento', aggfunc='sum')
display(pivot_table)
```

```python
#tabla dinamica con multiples columnas
pivot_table_mult = df.pivot_table(values='Salario', index=['Departamento','Empleado'],columns='Experiencia', aggfunc='sum')
display(pivot_table_mult)
```

```python
#crosstabulation, sirve para crear tablas de contingencia, que muestra cualidades
cross_tab = pd.crosstab(df['Departamento'], df['Empleado'])
display(cross_tab)
#osea contabiliza los valores
```

```python
#tablas de contingencia con funciones de agregacion
cross_tab_agg = pd.crosstab(df['Departamento'], df['Empleado'], values=df['Salario'], aggfunc='sum', margins=True)
display(cross_tab_agg)
```

```python
#tablas de contingencia normalizadas
cross_tab_norm = pd.crosstab(df['Departamento'], df['Empleado'], normalize='index')
display(cross_tab_norm)
```
